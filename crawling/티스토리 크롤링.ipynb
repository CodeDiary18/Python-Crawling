# 티스토리 크롤링
### chromedrive.exe의 위치는 이 주피터 노트북이 있는 디렉토리 안에 있어야 함

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
import sys
```

```python
query_text = input('크롤링할 키워드를 입력하세요 : ')
# f_name = input('검색 결과를 저장할 파일경로와 이름을 입력하세요 : ')
```

```python
# chrome 드라이버를 사용해서 웹 브라우저를 실행
path="./chromedriver.exe"
driver=webdriver.Chrome(path)
driver.get("여기에 티스토리 주소 넣기") # 티스토리
time.sleep(2) # 페이지가 모두 열릴 때 까지 2초 기다리기
```

```python
# 검색창에 검색어를 입력
driver.find_element_by_class_name("btn_search").click()
element=driver.find_element_by_class_name('tf_search')
element.send_keys(query_text)

# 검색 실행
element.send_keys(Keys.ENTER)
```

```python
# 현재 페이지에 있는 내용을 화면에 출력
while True:
    time.sleep(1)
    full_html=driver.page_source
    time.sleep(2) # 페이지가 모두 열릴 때 까지 2초 기다리기
    soup = BeautifulSoup(full_html, 'html.parser')
    content_list=soup.find_all('strong', class_='tit_post')
    for i in content_list:
        print(i.text.strip())
        print('\n')
    next_is=soup.find_all('a', class_="btn_next no-more-next")
    if len(next_is)!=0:
        break
    driver.find_element_by_class_name("btn_next").click()
```

```python
driver.close()
```
